```python
import requests

# Cache to store previously fetched news articles
cached_articles = []

def get_news(api_key, source):
    url = f"https://newsapi.org/v2/top-headlines?sources={source}&apiKey={api_key}"
    response = requests.get(url)
    data = response.json()

    if data["status"] == "ok":
        articles = data["articles"]
        new_articles = []

        # Check against cache to fetch only new articles
        for article in articles:
            if article not in cached_articles:
                new_articles.append(article)

        if new_articles:
            for article in new_articles:
                title = article["title"]
                author = article["author"]
                description = article["description"]
                print(f"📰 Title: {title}")
                print(f"👤 Author: {author}")
                print(f"💡 Description: {description}")
                print("")

            # Update the cache with newly fetched articles
            cached_articles.extend(new_articles)
        else:
            print("No new articles found.")

# Replace 'YOUR_API_KEY' with your actual News API key
api_key = 'e7736d9b33154b0a8f2e9193789c95f9'

# Specify the news source you want to fetch headlines from
source = 'bbc-news'

print("🗞️ Fetching Latest News Headlines 🗞️")
print("")

get_news(api_key, source)
```
